**Linear Regression**

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
# 현재 실습하고 있는 파이썬 코드를 재실행해도 다음에도 같은 결과가 나오도록 랜덤 시드(random seed) 설정
torch.manual_seed(1)
```

**Training**

```python
# 데이터셋 생성 (강의자료 참조하여 구현하시오.)
x_train = torch.FloatTensor([[1], [2], [3]])
y_train = torch.FloatTensor([[4], [5], [6]])

# 모델 초기화 (강의자료 참조하여 구현하시오.)
W = torch.zeros(1, requires_grad=True)
b = torch.zeros(1, requires_grad=True)

# optimizer 설정 (강의자료 참조하여 구현하시오.) 
optimizer = optim.SGD([W, b], lr=0.01)

nb_epochs = 1000 # 원하는만큼 경사 하강법을 반복
for epoch in range(nb_epochs + 1):
    # 가설 및 손실함수 (강의자료 참조하여 구현하시오.)
    output = W * x_train + b
    loss = torch.mean((output - y_train)**2)

    # 경사하강법 (강의자료 참조하여 구현하시오.)
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    # 100번마다 로그 출력
    if epoch % 100 == 0:      
        print('Epoch {:4d}/{} W: {:.3f}, b: {:.3f} Cost: {:.6f}'.format(
            epoch, nb_epochs, W.item(), b.item(), loss.item()
        ))
```

**Inference**

```python
#아직 결과값을 모르는 데이터에 대한 추론 (강의자료 참조하여 구현)
test_data = torch.FloatTensor([[4]])
prediction = W * test_data + b
print(prediction)
```

**과제**

아래 그림과 같은 학습 데이터에 대해서 선형회귀를 수행하시오.
(독립 변수로는 공부한 시간을 사용하고, 예측하고자 하는 결과값으로는 점수를 사용할 것. ex) 6 -> 53, 7->60)

그리고 학습된 모델을 이용하여 8.2시간 공부한 Brady라는 학생의 점수를 예측해보시오.

image.png

```python
x_train2 = torch.FloatTensor([6, 7, 6.5, 8, 6.6, 8.1, 6.8, 6.9, 7.3, 6.9]).view(-1, 1)
y_train2 = torch.FloatTensor([53, 60, 56, 79, 58, 85, 70, 56, 69, 76]).view(-1, 1)

W2 = torch.zeros(1, requires_grad=True)
b2 = torch.zeros(1, requires_grad=True)

optimizer2 = optim.SGD([W2, b2], lr=0.01)

num_epoch = 10000
for epoch in range(1, num_epoch + 1):
    output2 = W2 * x_train2 + b2
    loss2 = torch.mean((output2 - y_train2)**2)

    optimizer2.zero_grad()
    loss2.backward()
    optimizer2.step()

    if epoch % 1000 == 0:      
        print('Epoch {:4d}/{} W: {:.3f}, b: {:.3f} Cost: {:.6f}'.format(
            epoch, num_epoch, W2.item(), b2.item(), loss2.item()
        ))
```

```python
test_data = torch.FloatTensor([8.2]).view(-1, 1)
predict = W2 * test_data + b2
print(predict)
```
