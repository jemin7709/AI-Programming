**numpy array 생성**



```python
import numpy as np

a = np.array([1, 2, 3])  # rank가 1인 배열 생성
print (type(a))            # 출력 "<type 'numpy.ndarray'>"
print (a.shape)           # 출력 "(3,)"
print (a[0], a[1], a[2])   # 출력 "1 2 3"
a[0] = 5                 # 요소를 변경
print (a)                  # 출력 "[5, 2, 3]"

b = np.array([[1,2,3],[4,5,6]])   # rank가 2인 배열 생성
print (b.shape)                     # 출력 "(2, 3)"
print (b[0, 0], b[0, 1], b[1, 0])   # 출력 "1 2 4"
```

**numpy array 인덱싱 및 슬라이싱**

```python
t = np.array([0., 1., 2., 3., 4., 5., 6.])

print('t[0] t[1] t[-1] = ', t[0], t[1], t[-1]) # 인덱스를 통한 원소 접근
print('t[2:5] t[4:-1]  = ', t[2:5], t[4:-1]) # [시작 번호 : 끝 번호]로 범위 지정을 통해 가져온다.
print('t[:2] t[3:]     = ', t[:2], t[3:]) # 시작 번호를 생략한 경우와 끝 번호를 생략한 경우
```

**numpy dtype**

```python
x = np.array([1, 2])  # Numpy가 자료형을 추측해서 선택
print (x.dtype)         # 출력 "int64"

x = np.array([1.0, 2.0])  # Numpy가 자료형을 추측해서 선택
print (x.dtype)             # 출력 "float64"

x = np.array([1, 2], dtype=np.int64)  # 특정 자료형을 명시적으로 지정
print (x.dtype)                         # 출력 "int64"
```

**numpy matrix operations**

```python
x = np.array([[1,2],[3,4]], dtype=np.float64)
y = np.array([[5,6],[7,8]], dtype=np.float64)

# 요소별 합; 둘 다 다음의 배열을 만듭니다
# [[ 6.0  8.0]
#  [10.0 12.0]]
print (x + y)
print (np.add(x, y))

# 요소별 차; 둘 다 다음의 배열을 만듭니다
# [[-4.0 -4.0]
#  [-4.0 -4.0]]
print (x - y)
print (np.subtract(x, y))

# 요소별 곱; 둘 다 다음의 배열을 만듭니다
# [[ 5.0 12.0]
#  [21.0 32.0]]
print (x * y)
print (np.multiply(x, y))

# 요소별 나눗셈; 둘 다 다음의 배열을 만듭니다
# [[ 0.2         0.33333333]
#  [ 0.42857143  0.5       ]]
print (x / y)
print (np.divide(x, y))

# 요소별 제곱근; 다음의 배열을 만듭니다
# [[ 1.          1.41421356]
#  [ 1.73205081  2.        ]]
print (np.sqrt(x))
```

```python
x = np.array([[1,2],[3,4]])
y = np.array([[5,6],[7,8]])

v = np.array([9,10])
w = np.array([11, 12])

# 벡터의 내적; 둘 다 결과는 219
print (v.dot(w))
print (np.dot(v, w))

# 행렬과 벡터의 곱; 둘 다 결과는 rank 1인 배열 [29 67]
print (x.dot(v))
print (np.dot(x, v))

# 행렬곱; 둘 다 결과는 rank 2인 배열
# [[19 22]
#  [43 50]]
print (x.dot(y))
print (np.dot(x, y))
```

**과제1**

아래 코드에 #으로 이루어진 5칸짜리 정사각형을 출력해주는 함수 rect가 있다.
rect를 변경하여 아래 그림과 같이 원하는 크기(5 이상)를 입력받고, 해당 크기에 해당하는 #로 이루어진 정사각형을 출력하도록 하여라.

lec04_1.png

```python
#Edit below codes
def rect(length):
  print ('#'* length)
  for i in range(0, length - 2):
    print ('#' + ' '* (length - 2) + '#')
  print ('#'*length)

rect(5)
rect(7)
rect(15)
```

**과제2**

앞서 구현한 rect 함수를 이용하여, 아래와 같이 주어진 크기(5이상 홀수)에 대해 모래시계 모양을 출력하는 sandglass 함수를 구현하시오.

[Hint] 모래시계 모양을 만들기 위해 한 쪽에 출력해야 하는 빈 칸 개수(0->1->2->...->2->1->0)
를 별도의 변수로 관리하고 for문을 이용해 모래시계 출력 구현

image.png

```python
#Implement here
def sandglass(length):
  n = 0
  s = length - 2
  c = True
  print ('#'* length)
  for i in range(0, (length - 2)):
    print('#' + ' ' * n, end='')
    print('*' * s, end='')
    print(' ' * n + '#')
    if(c):
      n += 1
      s -= 2
      if(s == 1):
        c = False
    else:
      n -= 1
      s += 2
  print ('#'*length)

sandglass(5)
sandglass(7)
sandglass(15)
```

**과제3**

다음과 같이 동작하는 클래스 Calculator를 작성하시오.

image.png

```python
#Implement here
class Calculator():
  def __init__(self, l):
    self.l = l
  def sum(self):
    print(sum(self.l))
  def avg(self):
    print(sum(self.l) / len(self.l))
  
cal1 = Calculator([1,2,3,4,5])
cal1.sum()
cal1.avg()
cal2 = Calculator([6,7,8,9,10])
cal2.sum()
cal2.avg()
```
