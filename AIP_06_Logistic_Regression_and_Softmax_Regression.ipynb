**Logistic Regression**

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
# 현재 실습하고 있는 파이썬 코드를 재실행해도 다음에도 같은 결과가 나오도록 랜덤 시드(random seed) 설정
torch.manual_seed(1)
```

**Model code**

```python
class LogisticRegression(nn.Module):
    def __init__(self):
        super().__init__()
        self.linear = nn.Linear(2, 1)
        self.sigmoid = nn.Sigmoid()

    def forward(self, x):
      return self.sigmoid(self.linear(x))
```

**Training**

```python
#구현 필요
x_train = torch.FloatTensor([[1, 2], [2, 3], [3, 1], [4, 3], [5, 3], [6,2]])
y_train = torch.FloatTensor([0, 0, 0, 1, 1, 1]).view(-1, 1)

model = LogisticRegression()
optimizer = torch.optim.SGD(model.parameters(), lr=1e-1)
nb_epochs = 1000
for epoch in range(nb_epochs+1):
    #구현 필요
    prediction = model(x_train)
    cost = F.binary_cross_entropy(prediction, y_train)

    optimizer.zero_grad()    
    cost.backward()
    optimizer.step()

    if epoch % 100 == 0:
      prediction_class = prediction >= torch.FloatTensor([0.5])
      correct_prediction = prediction_class.float() == y_train
      accuracy = correct_prediction.sum().item() / len(correct_prediction)
    # 100번마다 로그 출력
      #구현 필요
      print('Epoch {:4d}/{} Cost: {:.6f} Accuracy: {:2.2f}%'.format(
          epoch, nb_epochs, cost.item(), accuracy*100
      ))
```

**Inference**

```python
#구현 필요
test_data = torch.FloatTensor([[5, 5], [1, 4]])
prediction = model(test_data)
print (prediction)
```

**Softmax Regression**

**Model code**

```python
class SoftmaxRegression(nn.Module):
    def __init__(self):
        super().__init__()
        self.linear = nn.Linear(4, 3)

    def forward(self, x):
        return self.linear(x)
```

**Training**

```python
x_train = torch.FloatTensor([[6.3, 3.3, 6.0, 2.5], 
                             [5.8, 2.7, 5.1, 1.9], 
                             [7.1, 3.0, 5.9, 2.1], 
                             [5.1, 3.5, 1.4, 0.2], 
                             [4.9, 3.0, 1.4, 0.2], 
                             [4.7, 3.2, 1.3, 0.2],
                             [7.0, 3.2, 4.7, 1.4],
                             [6.4, 3.2, 4.5, 1.5],
                             [6.9, 3.1, 4.9, 1.5]])  
#구현 필요
# Class 0: virginica, Class 1: setosa, Class 2: versicolor
y_train = torch.LongTensor([0, 0, 0, 1, 1, 1, 2, 2, 2])

#구현 필요
model = SoftmaxRegression()
optimizer = torch.optim.SGD(model.parameters(), lr=1e-1)
nb_epochs = 1000
for epoch in range(nb_epochs+1):
    #구현 필요
    prediction = model(x_train)
    cost = F.cross_entropy(prediction, y_train)

    optimizer.zero_grad()    
    cost.backward()
    optimizer.step()

    if epoch % 100 == 0:
    # 100번마다 로그 출력
      #구현 필요
      correct_prediction = torch.argmax(prediction, 1) == y_train
      accuracy = correct_prediction.float().mean().item()
      print('Epoch {:4d}/{} Cost: {:.6f} Accuracy: {:2.2f}%'.format(
          epoch, nb_epochs, cost.item(), accuracy*100
      ))
```

**Testing**

```python
test_data = torch.FloatTensor([[6.3, 2.9, 5.6, 1.8], 
                               [6.5, 3.0, 5.8, 2.2],
                               [4.6, 3.1, 1.5, 0.2],
                               [5.0, 3.6, 1.4, 0.2],
                               [5.5, 2.3, 4.0, 1.3],
                               [6.5, 2.8, 4.6, 1.5]])  
# Class 0, 1, 2 두개씩 포함하고 있는 테스트 데이터셋

#구현 필요
prediction = F.softmax(model(test_data), 1)
print (prediction)
```

**Softmax Regression for MNIST dataset**

**Model code**

```python
class MNIST_SoftmaxRegression(nn.Module):
    def __init__(self):
        super().__init__()
        self.linear = nn.Linear(784, 10)

    def forward(self, x):
        return self.linear(x)
```

**Training**

```python
import torchvision.datasets as dsets
import torchvision.transforms as transforms
from torch.utils.data import DataLoader
import matplotlib.pyplot as plt
import random

# hyperparameters
#구현 필요
training_epochs = 15
batch_size = 100

# MNIST dataset
mnist_train = dsets.MNIST(root='../MNIST_data/',
                          train=True,
                          transform=transforms.ToTensor(),
                          download=True)

mnist_test = dsets.MNIST(root='../MNIST_data/',
                         train=False,
                         transform=transforms.ToTensor(),
                         download=True)

# dataset loader
#구현 필요
data_loader = DataLoader(mnist_train, batch_size=batch_size, shuffle=True)
criterion = nn.CrossEntropyLoss()
model = MNIST_SoftmaxRegression()
optimizer = torch.optim.SGD(model.parameters(), lr=0.1)

for epoch in range(training_epochs): # 앞서 training_epochs의 값은 15로 지정함.
    avg_cost = 0
    total_batch = len(data_loader)

    for idx, (x_train, y_train) in enumerate(data_loader):
        #구현 필요
        x_train = x_train.view(-1, 784)
        output = model(x_train)
        cost = criterion(output, y_train)

        optimizer.zero_grad()
        cost.backward()
        optimizer.step()

        avg_cost += cost / total_batch
    print('Epoch:', '%04d' % (epoch + 1), 'cost =', '{:.9f}'.format(avg_cost))

print('Learning finished')
```

```python
# 테스트 데이터를 사용하여 모델을 테스트한다.
with torch.no_grad(): # torch.no_grad()를 하면 gradient 계산을 수행하지 않는다.
    #구현 필요
    x_test = mnist_test.data.view(-1, 784).float()
    y_test = mnist_test.targets
    
    prediction = model(x_test)

    correct_prediction = torch.argmax(prediction, 1) == y_test
    accuracy = correct_prediction.float().mean()
    print('Accuracy:', accuracy.item())
```

```python
with torch.no_grad():    
    # MNIST 테스트 데이터에서 무작위로 하나를 뽑아서 예측을 해본다
    r = random.randint(0, len(mnist_test) - 1)
    x_single_data = mnist_test.data[r:r + 1].view(-1, 28 * 28).float()
    y_single_data = mnist_test.targets[r:r + 1]

    print('Label: ', y_single_data.item())
    single_prediction = model(x_single_data)
    print('Prediction: ', torch.argmax(single_prediction, 1).item())

    plt.imshow(mnist_test.data[r:r + 1].view(28, 28), cmap='Greys', interpolation='nearest')
    plt.show()
```
