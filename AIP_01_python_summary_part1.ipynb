**실습문제 4**

while문을 활용하여 1부터 1000까지의 자연수 중 3의 배수의 합을 구해서 출력하시오.

```python
#Implement here
a = 1
total = 0
while a <= 1000:
  if a % 3 == 0:
    total += a
  a += 1
print(total)
```

**실습문제 5**

사용자가 -1을 입력하기 전까지 입력하는 모든 수를 리스트에 저장했다가, -1을 입력하면 

1) 지금까지 입력된 수를 리스트로 출력

2) 합계 출력

3) 숫자 개수 출력

을 수행하는 프로그램을 구현하시오.
(사용자는 -1 및 자연수만 입력한다고 가정)

<실행 예시>

image.png

```python
nums = []
while True:
  curr_num = int(input("0 이상의 숫자를 입력해주세요(-1은 종료): "))
  
  #Implement here
  if curr_num == -1:
    print("숫자 리스트: {}".format(nums))
    print("합계: {}".format(sum(nums)))
    print("개수: {}".format(len(nums)))
    break
  else:
    nums.append(curr_num)
```
